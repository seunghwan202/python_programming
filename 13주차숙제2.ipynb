```python
import requests
from bs4 import BeautifulSoup

def get_book_details(isbn):
    details = {
        'isbn': isbn,
        'author': '',
        'category': '',
        'page_count': '',
        'weight': '',
        'size': ''
    }

    
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser')

    
    prd_info = soup.find('a', attrs={'class': 'gd_name'})
    if prd_info is None:
        return details

  
    url = 'http://www.yes24.com' + prd_info['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')

   
    author_info = soup.find('span', attrs={'class': 'gd_auth'})
    details['author'] = author_info.get_text().strip() if author_info else ''

    
    category_info = soup.find('ul', attrs={'class': 'yesAlertLi'})
    details['category'] = category_info.get_text().strip() if category_info else ''

 
    prd_detail = soup.find('div', attrs={'id': 'infoset_specific'})
    prd_tr_list = prd_detail.find_all('tr')
    for tr in prd_tr_list:
        header_text = tr.find('th').get_text()
        if header_text == '쪽수, 무게, 크기':
            details_text = tr.find('td').get_text().split(',')
            details['page_count'] = details_text[0].strip()
            if len(details_text) > 1:
                details['weight'] = details_text[1].strip()
            if len(details_text) > 2:
                details['size'] = details_text[2].strip()
            break

    return details


example_isbn = 9791190090018
book_details = get_book_details(example_isbn)
print(book_details)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


def get_book_details(isbn):
    details = {
        'isbn': isbn,
        'author': '',
        'category': '',
        'page_count': '',
        'weight': '',
        'size': ''
    }

    
    url = 'http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser')

    
    prd_info = soup.find('a', attrs={'class': 'gd_name'})
    if prd_info is None:
        return details

    
    url = 'http://www.yes24.com' + prd_info['href']
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')

 
    author_info = soup.find('span', attrs={'class': 'gd_auth'})
    details['author'] = author_info.get_text().strip() if author_info else ''

    
    category_info = soup.find('ul', attrs={'class': 'yesAlertLi'})
    details['category'] = category_info.get_text().strip() if category_info else ''

    # Extracting other details (page count, weight, size)
    prd_detail = soup.find('div', attrs={'id': 'infoset_specific'})
    prd_tr_list = prd_detail.find_all('tr')
    for tr in prd_tr_list:
        header_text = tr.find('th').get_text()
        if header_text == '쪽수, 무게, 크기':
            details_text = tr.find('td').get_text().split(',')
            details['page_count'] = details_text[0].strip()
            if len(details_text) > 1:
                details['weight'] = details_text[1].strip()
            if len(details_text) > 2:
                details['size'] = details_text[2].strip()
            break

    return details


isbn_list = [9791190090018, 9788965702828]  
books_details = [get_book_details(isbn) for isbn in isbn_list]


books_df = pd.DataFrame(books_details)

# Adding two new columns: 'Author' and 'Category'
books_df['Author'] = books_df['author']
books_df['Category'] = books_df['category']

print(books_df)
```
